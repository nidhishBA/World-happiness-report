```python
all_countries = df["Country name"].value_counts().reset_index()["index"].tolist()
all_countries_pop = pop["Country Name"].value_counts().reset_index()["index"].tolist()

del_country = []
for x in all_countries:
    if x not in all_countries_pop:
        del_country.append(x)
del_country
```

```python
pop_df = df[["Log GDP per capita","Life Ladder","Country name","year","Healthy life expectancy at birth","Freedom to make life choices","Generosity","Perceptions of corruption","region"]].copy()
pop_df.head()
```

```python
pop_df=pop_df[~pop_df["Country name"].isin(del_country)]
pop_df = pop_df[~pop_df["year"].isin([2006,2005,2007,2018,2019,2020,2021])]
pop_dict = {x : {} for x in range(2008,2018)}
for i in range(len(pop)):
    if(pop["Year"][i] in range(2008,2018)):
        pop_dict[pop["Year"][i]][pop["Country Name"][i]] = pop["Count"][i]
```

```python
population = []
for i in pop_df.index:
    population.append(pop_dict[pop_df["year"][i]][pop_df["Country name"][i]])
pop_df["population"]= population
```

# Top ten and bottom ten countries analysis on basis Ladder score

```python
top_10_df = df2021.sort_values(by="Ladder score", ascending=False).head(10)

# Create a horizontal bar plot with points and text labels
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_df, x="Ladder score", y="Country name", hue="Regional indicator", palette="Set3")
plt.title("Top Ten Happiest Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Countries")

for index, row in top_10_df.iterrows():
    plt.text(row["Ladder score"], index, f'{row["Ladder score"]:.2f}', va='center', fontsize=12)

plt.legend(title="Region", loc="upper right")
plt.show()
```

```python
bottom_10_df = df2021.sort_values(by="Ladder score").head(10)

# Create a horizontal bar plot with points and text labels
plt.figure(figsize=(10, 6))
sns.barplot(data=bottom_10_df, x="Ladder score", y="Country name", hue="Country name", palette="Set3")
plt.title("Bottom Ten Happiest Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Countries")

for index, row in bottom_10_df.iterrows():
    plt.text(row["Ladder score"], index, f'{row["Ladder score"]:.2f}', va='center', fontsize=12)

plt.legend(title="Region", loc="upper right")
plt.show()
```
